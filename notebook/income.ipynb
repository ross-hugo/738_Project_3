```python
from script.nn import NN
```

```python
#census data to determine the income of someone
census_df = pd.read_csv('data/adult_dataset/adult-all.csv', header=None)
census_df.replace('?', np.nan, inplace=True)
census_df.dropna(inplace=True)
census_df
```

```python
le = LabelEncoder()

labeled_df = census_df.copy()

for i in range(len(labeled_df.columns)):
    le.fit(labeled_df[i])
    labeled_df[i] = le.transform(labeled_df[i])

labeled_df
```

```python
train_df, test_df = labeled_df[:int(len(labeled_df) * 0.8)], labeled_df[int(len(labeled_df) * 0.8):]
print(len(train_df))
print(len(test_df))

train_input, train_output = train_df.drop(14, axis=1), train_df[len(train_df.columns) - 1]
test_input, test_output = test_df.drop(14, axis=1), test_df[len(test_df.columns) - 1]
```

```python
in_nodes = 14
out_nodes = 2

hidden_layers = [50, 50]

n = NN(in_nodes, out_nodes, hidden_layers)
```

```python
n_epochs = 8000
batch_size = 1024

learning_step = 1

# train the network
# n.train(X=train_input, y=train_output, epochs=n_epochs, batch_size=batch_size)
n.train(np.array(train_input), np.array(train_output), n_epochs, batch_size, learning_step)
```

```python
n.test(np.array(test_input), np.array(test_output))
```
